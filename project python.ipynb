```python
from bs4 import BeautifulSoup
import requests

url = BeautifulSoup('http://nepalstock.com/todaysprice', 'html.parser')
soup = requests.get(url)
soup
```

```python
soup = soup.text
soup
```

```python
soup = BeautifulSoup(soup, "lxml")
soup
```

```python
soup = soup.table
soup
```

```python
tags = soup.find_all('tr')
tags
```

```python
tags[1]
```

```python
tags[1].text
```

```python
for i in tags:
    print(i.text)
```

```python
data = []
for i in tags:
    data.append(i.text.split('\n')[1:-1])
data
```

```python
import csv
file = open('Nepse1.csv','w')
a = csv.writer(file)
for i in data:
    a.writerow(i)
```

```python
output_data = [i for i in data if i[0] != ""]
print(output_data)
```
